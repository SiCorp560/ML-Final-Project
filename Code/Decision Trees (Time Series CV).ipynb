```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
```

```python
df = pd.read_csv("mayoralElectionsv2.csv")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df.head(5)
```

```python
df = df[["year", "month", "candidate_count", "pid_est", "gender_est", "winner"]]
df["challenger_win"] = df["winner"].map({"win": 0, "lose": 1})
```

```python
df.isna().sum()
```

```python
df["pid_est"].fillna("Unknown", inplace=True)
df["pid_est"].replace({"D": "Democrat", "R": "Republican"}, inplace=True)

df["gender_est"].fillna("Unknown", inplace=True)
df["gender_est"].replace({"M": "Male", "F": "Female"}, inplace=True)
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df_sorted = df.sort_values(["year", "month"], ascending=[True, True])
df_sorted.head(5)
```

```python
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df[["candidate_count", "pid_est", "gender_est"]]
Y = df["challenger_win"]
```

```python
ohe = OneHotEncoder()
std = StandardScaler()
ct = make_column_transformer(
    (ohe, ["pid_est", "gender_est"]),
    (std, ["candidate_count"]),
     remainder = 'passthrough'
)
model = DecisionTreeClassifier()
tscv = TimeSeriesSplit(n_splits=5)

pipe = make_pipeline(ct, model)

scores = cross_val_score(pipe, X, Y, cv=tscv, scoring='accuracy')

print("Cross-validated scores (time-aware):", scores)
print("\nMean accuracy:", scores.mean())
```
