# Read Data

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer, KNNImputer

encoder = LabelEncoder()
imputer = SimpleImputer(strategy='mean')
knn_imputer = KNNImputer()
```

```python
df = pd.read_csv("NEWNEW.csv")
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
df.count()
```

# Remove/Impute N/A Values

```python
df.isna().sum()
```

### Time

```python
df.loc[df["month"].isna()].head(5)
```

```python
df["month"].unique()
```

```python
df = df.drop("month", axis=1)
#df["month"] = df["month"].fillna(1)
```

### Political Party

```python
df.loc[df["pid_est"].isna()].head(5)
```

```python
df["pid_est"].unique()
```

```python
df["pid_est_knn"] = encoder.fit_transform(df["pid_est"])
df["pid_est_knn"] = df["pid_est_knn"].replace(2, np.nan)
df["pid_est_knn"] = knn_imputer.fit_transform(df[["pid_est_knn"]])
df["pid_est_knn"].unique()
```

```python
df["pid_est"] = df["pid_est"].fillna("unknown")
df["pid_est"].unique()
```

```python
df["prob_democrat"] = imputer.fit_transform(df[["prob_democrat"]])
df["prob_republican"] = imputer.fit_transform(df[["prob_republican"]])
```

### Gender

```python
df.loc[df["gender_est"].isna()].head(5)
```

```python
df["gender_est"].unique()
```

```python
df["gender_est_knn"] = encoder.fit_transform(df["gender_est"])
df["gender_est_knn"] = df["gender_est_knn"].replace(2, np.nan)
df["gender_est_knn"] = knn_imputer.fit_transform(df[["gender_est_knn"]])
df["gender_est_knn"].unique()
```

```python
df["gender_est"] = df["gender_est"].fillna("unknown")
df["gender_est"].unique()
```

```python
df["prob_male"] = imputer.fit_transform(df[["prob_male"]])
df["prob_female"] = imputer.fit_transform(df[["prob_female"]])
```

### Race

```python
df.loc[df["race_est"].isna()].head(5)
```

```python
df["race_est"].unique()
```

```python
df["race_est_knn"] = encoder.fit_transform(df["race_est"])
df["race_est_knn"] = df["race_est_knn"].replace(5, np.nan)
df["race_est_knn"] = knn_imputer.fit_transform(df[["race_est_knn"]])
df["race_est_knn"].unique()
```

```python
df["race_est"] = df["race_est"].fillna("unknown")
df["race_est"].unique()
```

```python
df["prob_black"] = imputer.fit_transform(df[["prob_black"]])
df["prob_white"] = imputer.fit_transform(df[["prob_white"]])
df["prob_hispanic"] = imputer.fit_transform(df[["prob_hispanic"]])
df["prob_asian"] = imputer.fit_transform(df[["prob_asian"]])
df["prob_other"] = imputer.fit_transform(df[["prob_other"]])
```

```python
df.isna().sum()
```

### Population

```python
df.loc[df["percent_women"].isna()].head(5)
```

```python
df = df.sort_values(["geo_name", "year"])
```

```python
df["percent_women"] = df.groupby("geo_name")["percent_women"].ffill()
df["percent_white"] = df.groupby("geo_name")["percent_white"].ffill()
df["percent_black"] = df.groupby("geo_name")["percent_black"].ffill()
df["percent_hispanic"] = df.groupby("geo_name")["percent_hispanic"].ffill()
df["percent_asian_american"] = df.groupby("geo_name")["percent_asian_american"].ffill()
```

### Contributor Ideology Score

```python
df = df.drop("bonica.cid", axis=1)
```

```python
df["contributor.cfscore"] = df.groupby("pid_est")["contributor.cfscore"].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isna().sum()
```

```python
df = df.dropna()
df.count()
```
