```python
import pandas as pd
```

```python
df = pd.read_csv("Mayoral Elections Cleaned.csv")
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X = df[["year", "month", "candidate_count", "pid_est", "gender_est",
        "race_est", "percent_women", "percent_white", "percent_black",
        "percent_hispanic", "percent_asian_american"]]
Y = df["challenger_win"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=69)
```

```python
ohe = OneHotEncoder()

ct = make_column_transformer(
    (ohe, ["month", "pid_est", "gender_est", "race_est"]),
     remainder = 'passthrough'
)

model = DecisionTreeClassifier()

pipeline = make_pipeline(ct, model)
pipeline.fit(X_train, Y_train)
```

```python
Y_pred = pipeline.predict(X_test)

accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
