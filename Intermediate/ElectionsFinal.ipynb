```python
# read in the data 

import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
#Read in elections data
candidates = pd.read_csv("c:/Users/zspba/OneDrive - andrew.cmu.edu/Desktop/ML_temp/ledb_candidatelevel.csv")
candidates.shape
```

```python
#Read in city demographics data
cityDem = pd.read_csv("c:/Users/zspba/OneDrive - andrew.cmu.edu/Desktop/ML_temp/cities_historical_demographics.csv")
cityDem.shape
```

```python
#Link all unique elections to the city demographic data
candidates.columns

candidates.dtypes
```

```python
candidates.head()
```

```python
#Select most relevant cols
df = candidates[['contest', 'office_consolidated','fips', 'geo_name', 'year', 'month', 'votes', 'vote_share', 'incumbent', 'winner','n_winners', 'prob_democrat', 'prob_republican', 'pid_est','prob_male', 'prob_female','gender_est', 'prob_black', 'prob_white', 'prob_hispanic','prob_asian','prob_other','race_est' ]]

#Only keep mayoral incumbent election candidates
df = df.loc[(df['incumbent'] == 1) & (df['office_consolidated'] == "Mayor")] 
df['fips'] = df['fips'].astype(int)

#delete the place_fips column
df = df.drop('office_consolidated', axis = 1)

df.head(15)
```

```python
cityDem.head(15)
cityDem.dtypes
```

```python
#join df with city demographic data
df = pd.merge(df, cityDem, left_on = ['fips', 'year'], right_on=['place_fips', 'year'], how = 'left')

#delete the place_fips column
df = df.drop('place_fips', axis = 1)

df.head(15)
```

```python
temp = candidates.groupby('contest').agg(candidate_count=('ledb_candid', 'count'))
temp = temp.reset_index()
temp
```

```python
#Merge temp and df

df = pd.merge(df, temp, on = 'contest', how = 'left')
df.head(30)
```

```python
#write to csv
df.to_csv('mayoralElections.csv', index=False)
```
